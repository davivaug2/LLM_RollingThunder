```python
# new way
# put all pdfs into1 folder , right now I am doing this manually
pdf_folder_path = "/content/folder1"
#  PyPDFDirectoryLoader different than  PyPDFLoader
from langchain.document_loaders import PyPDFDirectoryLoader
loader = PyPDFDirectoryLoader(pdf_folder_path)
# Load Documents and split into chunks. Chunks are returned as Documents.
# like pages
docs = loader.load_and_split()
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=150,
    chunk_overlap=15,
    separators=['\n\n', '\n', '(?=>\. )', ' ', '']
    )

# split pages into text chunks
texts = text_splitter.split_documents(docs)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

#embedding text
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2',model_kwargs={'device': 'cuda'})
```

```python
from langchain.vectorstores import FAISS

vectorstore = FAISS.from_documents(texts, embeddings)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftConfig, PeftModel
import transformers
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
new_trained_model_name = "jellyconsumer/new_model_falcon_large"
new_peft_model_name = "jellyconsumer/falcon-7b-instruct-sharded_large_peft"

old_trained_model_name = "vilsonrodrigues/falcon-7b-instruct-sharded"
model = AutoModelForCausalLM.from_pretrained(old_trained_model_name, quantization_config=bnb_config, trust_remote_code=True) #LLM Model
tokenizer = AutoTokenizer.from_pretrained(old_trained_model_name)



model.load_adapter(new_peft_model_name)
```

```python
from transformers import pipeline

pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens=1024,
                do_sample=True,
                top_k=10,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
from transformers import pipeline

pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens=1024,
                do_sample=True,
                top_k=10,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
from langchain import HuggingFacePipeline

llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
from langchain.chains import RetrievalQA

qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff",
                                 retriever=vectorstore.as_retriever(search_kwargs={"k": 3}),
                                 return_source_documents=True,
                                 verbose=False,
)
```

```python
query = "What is a technical order?"
query2 = "What is TODPG?"
query3 = "Who is responsible for specific aircraft nuclear weapon system TOs?"
# Send question as a query to qa chain
result = qa({"query": query3})
```

```python
from flask import Flask, request, jsonify, Response
from flask_ngrok import run_with_ngrok
from flask_cors import CORS
import json

class JSON_OBJECT:
    def toJSON(self):
        return json.dumps(self, default=lambda o: o.__dict__,
            sort_keys=True, indent=4)

#start the server with flask, ngrok and CORS
connect = Flask(__name__)
run_with_ngrok(connect)
CORS(connect)


@connect.route("/test", methods=["GET", "POST"])
def ask_question():
    #recieve the data requested from the website and decode it
    data = (request.data).decode("utf-8")

    #submit query
    result = qa(data)

    #query result => json
    answer = JSON_OBJECT()
    answer.result = result['result']
    answer.query = result['query']

    #return Response back to website
    return Response(answer.toJSON(), mimetype="application/json")

if __name__ == '__main__':
    connect.run()
```

```python
class Document:
    def __init__(self, page_content, metadata):
        self.page_content = page_content
        self.metadata = metadata

def print_result(result):
    pdf_name = result['source_documents'][0].metadata.get('source', '').split('/')[-1]

    # Print query, results, source documents header
    print("Query:", result.get('query', ''))
    print("\nResult:", result.get('result', ''))
    print("\nSource Documents (from", pdf_name, "):")

    for document in result.get('source_documents', []):
        print(f"  - Page {document.metadata.get('page', '')}: {document.page_content}")

print_result(result)
```
